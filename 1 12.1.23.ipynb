```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

DataLoop = pd.read_csv(r'C:\Users\xlbq78\OneDrive - Durham University\Rheometer Results\12.1\4.csv')
```

```python
freq = [0.2, 0.4, 0.8, 1.6, 3.2]

# Create a dictionary to hold data for each frequency
data_dict = {}
    
# Loop through the frequencies and create a variable for each one
for i in freq:
    data_key = f"{i}"
    data_dict[data_key] = DataLoop.loc[DataLoop['Frequency(Hz)'] == i]

# Create a list of markers for different frequencies
markers = ['o', 's', '^', 'D', 'v']

# Initialize a figure and axis
fig, ax1 = plt.subplots()

# Loop through the data_dict to plot the data
for i, data_key in enumerate(data_dict.keys()):
    data = data_dict[data_key]
    
    # Plot GePa with blue color and a marker on a logarithmic scale on the primary y-axis
    ax1.semilogy(data['Time (sequence)(s)'], data['Shear modulus (elastic component)(Pa)'],
                label=f'GePa, {data_key}Hz', color='blue', marker=markers[i])
    
    # Plot GvPa with red color and a marker on a logarithmic scale on the primary y-axis
    ax1.semilogy(data['Time (sequence)(s)'], data['Shear modulus (viscous component)(Pa)'],
                label=f'GvPa, {data_key}Hz', color='red', marker=markers[i])

# Set labels for x and primary y axes
ax1.set_xlabel('Time (sequence)(s)')
ax1.set_ylabel('Modulus (Pa)',fontsize=30)
#Time Solution Added/ Pump Started
plt.axvline(x=1200, color = 'orange')

# Create a secondary y-axis for Phase Angle (°)
ax2 = ax1.twinx()

# Loop through the data_dict to plot Phase Angle with green color and a marker on a logarithmic scale
for i, data_key in enumerate(data_dict.keys()):
    data = data_dict[data_key]
    
    ax2.plot(data['Time (sequence)(s)'], data['Phase angle(°)'],
                label=f'Phase Angle, {data_key}Hz', color='green', linestyle='--', marker=markers[i])

# Set labels for secondary y-axis
ax2.set_ylabel('Phase Angle (°)')

# Place the legends below the plot using bbox_to_anchor

ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.192),
          ncol=5, fancybox=True, shadow=True)

ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.262),
          ncol=5, fancybox=True, shadow=True)

plt.title("Time Evolution 1mm Gap Size")

fig.set_size_inches(25, 10)

caption_text = "1mm gap size, 0.1M CaCl2,0.25M~ Alginate,  5 rpm pump"
plt.figtext(0.5, -0.02, caption_text, wrap=True, ha="center", fontsize=30)

# Show the plot
plt.show()
```

```python
dataselection = DataLoop.iloc[454:463]
dataselection2 = DataLoop.iloc[464:473]

fig, ax1 = plt.subplots(figsize=(10, 6))

x = dataselection["Frequency(Hz)"]
y1 = dataselection["Shear modulus (viscous component)(Pa)"]
y2 = dataselection["Shear modulus (elastic component)(Pa)"]
y3 = dataselection2["Shear modulus (viscous component)(Pa)"]
y4 = dataselection2["Shear modulus (elastic component)(Pa)"]

ax1.semilogy(x, y1, label="Shear modulus (viscous component) forward", color='blue', marker='o', linestyle='--')
ax1.semilogy(x, y2, label="Shear modulus (elastic component) forward", color='red', marker='s', linestyle='--')
ax1.semilogy(x, y3, label="Shear modulus (viscous component) backward", color='darkblue', marker='o', linestyle='--')
ax1.semilogy(x, y4, label="Shear modulus (elastic component) backward", color='maroon', marker='s', linestyle='--')
ax1.set_xscale('log')
ax1.set_xlabel("Frequency (Hz)")  # Corrected label string
ax1.set_ylabel("Shear Modulus (Pa)")
ax1.tick_params(axis='y', labelcolor='blue')

# Create a twinned y-axis for Phase angle
ax2 = ax1.twinx()
y5 = dataselection["Phase angle(°)"]
y6 = dataselection2["Phase angle(°)"]
ax2.plot(x, y5, label="Phase angle(°) Forward", color='lime', marker='^')
ax2.plot(x, y6, label="Phase angle(°) Reverse", color='darkgreen', marker='^')
ax2.set_ylabel("Phase angle(°)", color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim([0, 90])  # Set the y-axis limit for Phase angle

# Combine legends
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.192),
          ncol=5, fancybox=True, shadow=True)
ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.262),
          ncol=5, fancybox=True, shadow=True)

fig.set_size_inches(25, 10)

plt.title("Frequency Table 1mm Gap Size")

caption_text = "1mm gap size, 0.1M CaCl2,0.25M~ Alginate,  5 rpm pump"
plt.figtext(0.5, 0.01, caption_text, wrap=True, ha="center", fontsize=10)

plt.show()
```
